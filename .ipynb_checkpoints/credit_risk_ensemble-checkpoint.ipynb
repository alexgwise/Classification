## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
# added cell. Perform an integer encoding of the text columns
encoded_df = pd.get_dummies(df, columns=["home_ownership", "verification_status", "issue_d", "pymnt_plan", "initial_list_status", "next_pymnt_d", "application_type", "hardship_flag", "debt_settlement_flag"])
encoded_df.head()
```

## Split the Data into Training and Testing

```python
# Create our features
X = encoded_df.drop(columns="loan_status")
# Create our target
y = encoded_df["loan_status"]
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```
